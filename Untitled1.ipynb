```python
# Load all needed stock price data into cache
stock_data_cache = {}
tickers_to_load = df['ticker'].unique().tolist()
if 'SPY' not in tickers_to_load:
 tickers_to_load.append('SPY')
for ticker in tickers_to_load:
 with open(stock_files[ticker], 'rb') as f:
  print(f)
  try:
    data = pickle.load(f)
  except EOFError:
    print(f"Error loading data for {ticker}. Skipping...")
    df = df[df['ticker'] != ticker]
    continue
  price_df = data['price']
  price_df['Date'] = pd.to_datetime(price_df['Date'], errors='coerce')
  price_df['Date'] = price_df['Date'].dt.tz_localize(None)
  price_df.sort_values('Date', inplace=True)
  price_df.set_index('Date', inplace=True)
  stock_data_cache[ticker] = price_df


print(f"Loaded price data for {len(stock_data_cache)} tickers into cache.")
```

```python
def compute_trade_features(df):
  prior_30d_ret = []; prior_30d_vol = []
  prior_90d_ret = []; prior_90d_vol = []
  prior_180d_ret = []; prior_180d_vol = []
  rsi_list = []
  avg_vol_30d = []

  for idx, row in df.iterrows():
    ticker = row['ticker']
    tx_date = row['tx_date']
    file_date = row['file_date']
    price_df = stock_data_cache.get(ticker)
    if price_df is None:
      prior_30d_ret.append(np.nan); prior_30d_vol.append(np.nan);
      prior_90d_ret.append(np.nan); prior_90d_vol.append(np.nan);
      prior_180d_ret.append(np.nan); prior_180d_vol.append(np.nan);
      rsi_list.append(np.nan); avg_vol_30d.append(np.nan);
      continue
    def past_return_vol(end_date, days):
      start_date = end_date - pd.DateOffset(days=days)
      hist = price_df.loc[start_date: end_date, 'Close']
      if len(hist) < 2:
        return 0.0, 0.0
      ret = hist[-1] / hist[0] - 1
      vol = hist.pct_change().std()
      return ret, vol
    ret30, vol30 = past_return_vol(tx_date, 30)
    prior_30d_ret.append(ret30)
    prior_30d_vol.append(vol30)
    ret90, vol90 = past_return_vol(tx_date, 90)
    prior_90d_ret.append(ret90)
    prior_90d_vol.append(vol90)
    ret180, vol180 = past_return_vol(tx_date, 180)
    prior_180d_ret.append(ret180)
    prior_180d_vol.append(vol180)

    hist14 = price_df.loc[file_date - pd.DateOffset(days=14): file_date, 'Close']
    if len(hist14) < 2:
      rsi_list.append(np.nan)
    else:
      delta = hist14.diff().dropna()
      up = delta.clip(lower=0); down = -delta.clip(upper=0)
      avg_gain = up.mean(); avg_loss = down.mean()
      if avg_loss == 0:
        rsi = 100.0
      else:
        rs = avg_gain / avg_loss
        rsi = 100 - (100 / (1 + rs))
      rsi_list.append(rsi)

    if 'Volume' in price_df.columns:
      vol_hist = price_df.loc[file_date - pd.DateOffset(days=30): file_date, 'Volume']
      avg_vol = vol_hist.mean() if len(vol_hist) > 0 else np.nan
    else:
      avg_vol = np.nan
    avg_vol_30d.append(avg_vol)

  df['prior_1m_return'] = prior_30d_ret
  df['prior_1m_volatility'] = prior_30d_vol
  df['prior_3m_return'] = prior_90d_ret
  df['prior_3m_volatility'] = prior_90d_vol
  df['prior_6m_return'] = prior_180d_ret
  df['prior_6m_volatility'] = prior_180d_vol
  df['RSI_14d'] = rsi_list
  df['avg_vol_30d'] = avg_vol_30d
  df['momentum_factor'] = df['prior_1m_return'] * df['prior_1m_volatility'] # short-term  momentum vs risk
  df['log_tx_estimate'] = np.log1p(df['tx_estimate']) # log of transaction size
  df['relative_tx_size'] = df['tx_estimate'] / (df['prior_1m_volatility'] + 1e-6) # size relative to  volatility

  return df
df = compute_trade_features(df)
```

```python
initial_capital = 100000.0
capital = initial_capital
positions = []
current_idx = 0
trade_returns = []
trades = []
signals['file_date'] = pd.to_datetime(signals['file_date'])
unique_dates = sorted(signals['file_date'].dt.date.unique())
for current_date in unique_dates:
    for pos in positions.copy():
        if pos['exit_date'].date() <= current_date:
            ticker = pos['ticker']
            exit_price = stock_data_cache[ticker]['Close'].reindex([pos['exit_date']], method='ffill')[0]
            proceeds = pos['shares'] * exit_price
            capital += proceeds
            positions.remove(pos)
            trade_returns.append((exit_price - pos['entry_price']) / pos['entry_price'])
            trades.append(pos)
    todays_signals = signals[signals['file_date'].dt.date == current_date]
    if len(todays_signals) > 0:
        inv_vol = 1.0 / (todays_signals['prior_3m_volatility'] + 1e-6)
        weights = todays_signals['pred_prob'] * inv_vol
        weights = weights / weights.sum()
        for i, trade in todays_signals.iterrows():
            alloc = capital * weights.loc[i]
            ticker = trade['ticker']
            file_date = trade['file_date']
            price_series = stock_data_cache[ticker]
            next_idx = price_series.index.searchsorted(file_date + pd.Timedelta(days=1), side='left')
            if next_idx >= len(price_series):
                continue
            entry_date = price_series.index[next_idx]
            entry_price = price_series.loc[entry_date, 'Open'] if 'Open' in price_series.columns else price_series.iloc[next_idx]['Close']
            shares = alloc / entry_price
            capital -= alloc
            exit_date = entry_date + pd.DateOffset(days=90)
            if exit_date > price_series.index[-1]:
                exit_date = price_series.index[-1]
            positions.append({
                'entry_date': entry_date,
                'entry_price': entry_price,
                'ticker': ticker,
                'shares': shares,
                'exit_date': exit_date,
                'alloc_cost': alloc,
                'trade': trade
            })

for pos in positions:
    ticker = pos['ticker']
    exit_price = stock_data_cache[ticker]['Close'].reindex([pos['exit_date']], method='ffill')[0]
    proceeds = pos['shares'] * exit_price
    capital += proceeds

final_value = capital
profit_factor = final_value / initial_capital
CAGR = (profit_factor ** (1/(len(unique_dates)/252*12)) - 1)
print(f"Final capital: ${final_value:,.2f}")
```

```python
# Calculate CAGR and Sharpe Ratio for the strategy
val_df['file_date'] = pd.to_datetime(val_df['file_date'])
val_years = (val_df['file_date'].dt.date.max() - val_df['file_date'].dt.date.min()).days / 365.0
print(val_years)
strategy_CAGR = (final_value / initial_capital) ** (1/val_years) - 1
# Compute monthly returns for Sharpe (from the trade_returns we collected during simulation if available)
# Here we can approximate monthly returns from profit_factor:
monthly_returns = (profit_factor ** (1/(val_years*12))) - 1  # average monthly return

# Ensure index is in datetime format
stock_data_cache['SPY'].index = pd.to_datetime(stock_data_cache['SPY'].index)

# Resample to monthly data
monthly_prices = stock_data_cache['SPY'].resample('M').last()  # End-of-month Close price

# Compute monthly returns using percentage change
spy_returns = monthly_prices['Close'].pct_change().dropna()

# Debugging
print(f"Mean Monthly Return: {spy_returns.mean()}")
print(f"Monthly Std Dev: {spy_returns.std(ddof=1)}")
print(f"Total Monthly Data Points: {len(spy_returns)}")

# Sharpe Ratio for Monthly Returns (Annualized with √12)
spy_sharpe = (spy_returns.mean() / spy_returns.std(ddof=1)) * np.sqrt(12)
print(f"SPY Monthly Sharpe Ratio: {spy_sharpe}")
# Debugging: Print mean and std
print(f"Mean SPY Monthly Return: {np.mean(spy_returns)}")
print(f"SPY Monthly Std Dev: {np.std(spy_returns, ddof=1)}")

# Sharpe ratio with monthly scaling
spy_sharpe = (np.mean(spy_returns) / np.std(spy_returns, ddof=1)) * np.sqrt(12)
print(f"SPY Monthly Sharpe Ratio: {spy_sharpe}")

sharpe = (np.mean(trade_returns) / np.std(trade_returns, ddof=0)) * np.sqrt(12)
# SPY performance over the same period:
spy_start_price = stock_data_cache['SPY']['Close'].reindex([val_df['file_date'].dt.date.min()], method='ffill')[0]
spy_end_price   = stock_data_cache['SPY']['Close'].reindex([val_df['file_date'].dt.date.max() + pd.DateOffset(days=90)], method='ffill')[0]
print(val_df['file_date'].dt.date.max())
print(val_df['file_date'].dt.date.min())
print(spy_start_price)

print(spy_end_price)
spy_CAGR = (spy_end_price / spy_start_price) ** (1/val_years) - 1

print(f"Strategy CAGR: {strategy_CAGR:.2%}, Sharpe Ratio: {sharpe:.2f}")
print(f"SPY CAGR over same period: {spy_CAGR:.2%}, Sharpe Ratio: {spy_sharpe:.2f}")
```
